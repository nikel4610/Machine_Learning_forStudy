## 2 이차 함수와 최솟값
이차 함수란 y가 x에 관한 이차식으로 표현되는 경우를 의미합니다. 

$$y=ax^2 + bx+ c \;\;\;\;(a \ne 0)$$

2차식의 예를 들자면 다음과 같다.
$$y= (x-1)(x-1) = x^2 -2x + 1$$

```python
x = np.linspace(-4, 6, 100)
a = 1
b = -2
c = 1

y = a*x**2 + b*x + c
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y={}x^2 + ({})x +({})$'.format(a, b, c))

plt.axhline(0,color='black')    
plt.axvline(0,color='black')    
plt.grid()
plt.show()
```

```python
x = np.linspace(-4, 6, 100)
a = 1
b = -2
c = 1

y = a*x**2 + b*x + c

plt.figure(figsize=(8,8))     # 그림 크기 지정

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y={}x^2 + ({})x +({})$'.format(a, b, c))

plt.axhline(0,color='black')    
plt.axvline(0,color='black')    
plt.grid()
plt.show()
```

## 편미분
미분과 더불어 딥러닝을 공부할 때 가장 자주 접하게 되는 또 다른 수학 개념은 바로 편미분입니다. 미분과 편미분 모두 '미분하라 = 기울기(변화량)를 구하라'는 의미에서는 다를 바가 없습니다. 그러나 여러 가지 변수가 식에 포함되어 있을 때, 모든 변수를 미분하는 것이 아니라 우리가 원하는 한 가지 변수만 미분하고 그 외의 변수는 상수 취급하는 것이 편미분입니다. 예를 들어 $f(x)=2x+1$와 같은 식은 변수가 하나 뿐이어서 미분하라는 의미에 혼란이 없습니다. 그러나 다음 식과 같이 변수가 두 개($x_1, x_2$)라면 둘 중 어떤 변수로 미분을 하라는 것인지 명확히 표현해 줘야 합니다. 예를 들어 $x_1$에 대해 미분하라고 하면 $x_1$이 변 할때 함수 f의 값이 얼마나 변하는지 순각 변화량을 계산하라는 것입니다. 

$$f(x_1, x_2) = x{_1}^2 +x_1 x_2 + 3$$

위 식을 $x_1$에 대해 (편)미분하면  
$$ \frac{\partial f(x_1, x_2)}{\partial x_1} = 2x_1 + x_2$$


```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')

X=np.arange(-5,5,0.25)
Y=np.arange(-5,5,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X,Y,Z,cmap='coolwarm',linewidth=0,antialiased=False)
wire = ax.plot_wireframe(X,Y,Z,color='r',linewidth=0.1)
fig.colorbar(surf,shrink=0.5,aspect=5)
fig.tight_layout()

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```

위 그래프에서 $x_1$에 대해 편미분 했다고 가정해봅시다. $x_2$는 고정된 상수라고 생각 하겠습니다. 예를 들어 $x_2=-4$라고 할 때 $x_1$ 변화에 따른 그래프의 모양을 상상할 수 있습니까? 그 그래프의 기울기가 보입니까? $x_2=-2$라고 할 때 $x_1$ 변화에 따른 그래프의 모양을 상상할 수 있습니까? 그 그래프의 기울기가 보입니까?

## 7. 로그와 로그 함수
 

$a$는 1이 아닌 양의 상수($a \neq 1, a>0$)라고 할 때 두 변수 $x$와 $y$ 사이에는 $a^y=x$인 관계가 있으면 $y$는 $a$를 밑으로 하는 $x$의 로그함수라고 한다. 식 (7-1)과 같이 표현합니다.  
(식. 7-1)
$$ y = \log_{a} x $$

$a$가 양의 상수이므로 a의 거듭제곱인 $x$는 항상 양수입니다. 그러므로 식(7-1)의 정의역은 $x>0$인 실수가 됩니다. $a>1$일 때, 로그 함수 식 (7-1) 아래와 같습니다.

[log 공식 몇 개]
$$ \log_{a}M^P = P\log_{a}M$$

$$\log_{a}M = \frac{\log_{b}M}{\log_{b}a}$$

식(7-1)를 그래프로 그려보겠습니다. $a>1$라고 가정하겠습니다. 여기서는 $a=2$라고 가정했습니다
$$ y = \log_{2}x = \frac{\log_{10}{x}}{\log_{10}2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2
x = np.linspace(0.001,4,100)
y = np.log10(x) / np.log10(a)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y=\log_{a}x$')
plt.grid()
plt.show()
```

그래프를 통해 확인할 수 있듯이,  a>1 일 때 로그 함수  $y=\log_{a}x$ 는 증가함수이고 그래프는  $x=1$ 일 때  $x$ 축과 만납니다.

또한  $a>1$인 경우에 $a$의 값이 작을수록  $x>1$  영역에서 로그 함수  $y=\log_{a}x $의 그래프가 더 가파르게 증가합니다.

```python
a1 = 2
x = np.linspace(0.001,4,100)
y1 = np.log10(x) / np.log10(a1)


a2 = 4
y2 = np.log10(x) / np.log10(a2)

plt.plot(x,y1, label='a = 2')
plt.plot(x,y2, label='a = 4')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y=\log_{a}x$')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

$0<a<1$일 때, 로그함수 $y=\log_{a}x$의 그래프를 다음과 같습니다.

```python
a = 0.5
x = np.linspace(0.001,4,100)
y = np.log10(x) / np.log10(a)

plt.plot(x,y, label='a=0.5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y=\log_{a}x$')
plt.grid()
plt.legend()
plt.show()
```

$x=1$일 때 $x$축을 지나고 감소함수임을 알 수 있습니다. 

```python
a1 = 0.25
x = np.linspace(0.001,4,100)
y1 = np.log10(x) / np.log10(a1)


a2 = 0.5
y2 = np.log10(x) / np.log10(a2)

plt.plot(x,y1, label='a = 0.25')
plt.plot(x,y2, label='a = 0.5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y=\log_{a}x$')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

위 그래프는 $a = 0.25, a = 0.5$일 때의 그래프입니다. $a$값이 더 큰 그래프가 더 빨리 감소하는 것을 알 수 있습니다.
