## 5. í”¼ë§ˆ ì¸ë””ì–¸ ë‹¹ë‡¨ë³‘ ì˜ˆì¸¡ ì‹¤í–‰
ì´ì œ í…ì„œí”Œë¡œì˜ ì¼€ë¼ìŠ¤ë¥¼ ì´ìš©í•´ì„œ ì˜ˆì¸¡ì„ ì‹¤í–‰í•´ ë´…ì‹œë‹¤. íŒë‹¤ìŠ¤ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ê¸° ë•Œë¬¸ì— iloc() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ ì†ì„± $X$ì™€ í´ë˜ìŠ¤ $y$ë¥¼ ê°ê° ì €ì¥í•©ë‹ˆë‹¤. ilocëŠ” ë°ì´í„° í”„ë ˆì„ì—ì„œ ëŒ€ê´„í˜¸ ì•ˆì— ì •í•œ ë²”ìœ„ë§Œí¼ ìš”ì†Œë¥¼ ê°€ì ¸ì™€ ì €ì¥í•˜ê²Œ í•©ë‹ˆë‹¤.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# pandas ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
import pandas as pd

# ê¹ƒí—ˆë¸Œì— ì¤€ë¹„ëœ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤. ì´ë¯¸ ì•ì—ì„œ ê°€ì ¸ì™”ìœ¼ë¯€ë¡œ ì£¼ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤. 5ë²ˆ ì˜ˆì œë§Œ ë³„ë„ ì‹¤í–‰ ì‹œ ì£¼ì„ í•´ì œ í›„ ì‹¤ìŠµí•˜ì„¸ìš”.
!git clone https://github.com/taehojo/data.git

# í”¼ë§ˆ ì¸ë””ì–¸ ë‹¹ë‡¨ë³‘ ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤. 
df = pd.read_csv('./data/pima-indians-diabetes3.csv')
```

```python
# ì„¸ë¶€ ì •ë³´ë¥¼ Xë¡œ ì§€ì •í•©ë‹ˆë‹¤.
X = df.iloc[:,0:8]
# ë‹¹ë‡¨ë³‘ ì—¬ë¶€ë¥¼ yë¡œ ì§€ì •í•©ë‹ˆë‹¤.
y = df.iloc[:,8]
```

ë‹¤ìŒê³¼ ê°™ì´ ëª¨ë¸ êµ¬ì¡°ë¥¼ ì„¤ì •í•©ë‹ˆë‹¤. ì´ì „ê³¼ ë‹¬ë¼ì§„ ì ì€ ì€ë‹‰ì¸µì´ í•˜ë‚˜ ë” ì¶”ê°€ë˜ì—ˆë‹¤ëŠ” ê²ƒì…ë‹ˆë‹¤. ê·¸ë¦¬ê³  ì¸µê³¼ ì¸µì˜ ì—°ê²°ì„ í•œ ëˆ„ì— ë³¼ ìˆ˜ ìˆê²Œ í•´ì£¼ëŠ” model.summary()ì´ ì¶”ê°€ ë˜ì—ˆìŠµë‹ˆë‹¤. model.summary()ì˜ ì‹¤í–‰ ê²°ê³¼ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.

```python
# ëª¨ë¸ì„ ì„¤ì •í•©ë‹ˆë‹¤.
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu', name='Dense_1'))
model.add(Dense(8, activation='relu', name='Dense_2'))
model.add(Dense(1, activation='sigmoid',name='Dense_3'))
model.summary()
```
